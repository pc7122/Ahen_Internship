```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.google.co.in/maps/@18.5247219,73.8249368,13z?hl=en-GB')
```

```python
search = driver.find_element(By.NAME, 'q')
search.send_keys("Driving School in bengalore")
search.send_keys(Keys.RETURN)
```

```python
lst = driver.find_element(By.XPATH, '/html/body/div[3]/div[9]/div[9]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div[1]')
len(lst.find_elements(By.CSS_SELECTOR, 'div[jstcache="984"]'))
```

```python
html = lst.get_attribute('innerHTML')
```

```python
print(BeautifulSoup(html, 'html.parser').prettify())
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
urls = []
for i in soup.find_all('a', attrs={'class': 'hfpxzc'}):
    urls.append(i.get('href'))
```

```python
import re
re.search('\d\s', '095611 47519').string
```

```python
import time

dic = {
    'name': [],
    'rating': [],
    'address': [],
    'phone': [],
}
```

```python
com = re.compile(r'\d{6}\s\d{5}')

for url in urls:
    driver.get(url)
    time.sleep(3)
    
    html = driver.find_element(By.CSS_SELECTOR, '#QA0Szd > div > div > div.w6VYqd > div.bJzME.tTVLSc > div > div.e07Vkf.kA9KIf > div > div').get_attribute('innerHTML')
    soup1 = BeautifulSoup(html, 'html.parser')
    
    dic['name'].append(soup1.find('h1').text.strip())
    dic['rating'].append(soup1.find_all('span')[5].text.strip())
    
    dic['address'].append(soup1.find_all('div', attrs={'class':'rogA2c'})[0].text.strip())
    
    flag = 0
    for i in soup1.find_all('div', attrs={'class':'rogA2c'})[:5]:
        phone = i.text.strip()
        if com.search(phone) != None:
            dic['phone'].append(phone)
            flag = 1
            break
    if not flag:
        dic['phone'].append('nan')
 
```
